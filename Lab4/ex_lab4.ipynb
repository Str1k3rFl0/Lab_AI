```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ex. 1
data_set = pd.read_csv('data.csv')
data_set
```

```python
# Ex. 2
### mai simplu!
player_over_40 = data_set[data_set["Age"] > 40]
print(players_over_40[["Name", "Age"]].head(10))
```

```python
# Ex. 3
best_u25_players = data_set[(data_set["Overall"] >= 85) & (data_set["Age"] < 25)]
print(best_u25_players[["Name", "Overall", "Age"]])
```

```python
# Ex. 4
sorted_players = data_set.sort_values(by="Skill Moves", ascending=False)
print(sorted_players[["Name", "Skill Moves"]])
```

```python
# Ex. 5
# contract_2021 = data_set[data_set["Contract Valid Until"] <= 2021]
# print(contract_2021[["Name", "Contract Valid Until"]])

### unele coloane folosesc datetime si atunci convertim toata coloana la datetime!!
data_set["Contract Valid Until"] = pd.to_datetime(data_set["Contract Valid Until"], errors="coerce")
data_set["Contract Year"] = data_set["Contract Valid Until"].dt.year
contract_until_2021 = data_set[data_set["Contract Year"] <= 2021]
print(contract_until_2021.head(20))
```

```python
# Ex. 6
print(f"Dimensiune ( nr. randuri si nr. coloane ): {data_set.shape}")
unique_players = data_set["Name"].nunique()
print("Numar jucatori unici:", unique_players)
```

```python
# Ex. 7
most_common_nationality = data_set["Nationality"].value_counts().head(5)
print(most_common_nationality)
```

```python
# Ex. 8
plt.figure()
plt.pie(most_common_nationality, labels=most_common_nationality.index)
plt.title("Top 5 nationalitati")
plt.show()
```

```python
# Ex. 9
nationality_average = data_set.groupby("Nationality")[["SprintSpeed", "Acceleration"]].mean()
print(nationality_average)
```

```python
# Ex. 10
data_set["Position"] = data_set["Position"].fillna("Unknown")
print("Verificare scurta:\n", data_set["Position"].isnull().sum())
```

```python
# Ex. 11
club_average_overall = data_set.groupby("Club")["Overall"].mean()
best_club = club_average_overall.idxmax()
best_value = club_average_overall.max()

print("Cel mai mare overall al unui club:", best_club)
print("Medie overall:", best_value)
```

```python
# Ex. 12
count = (data_set["Value"] > data_set["Wage"]).sum()
print("Numarul de jucatori cu valoare de transfer mai mare decat salaru:", count)
```

```python
# Ex. 14
data_set["Score"] = (
    0.3 * data_set["Overall"] +
    0.3 * data_set["Potential"] +
    0.2 * data_set["SprintSpeed"] +
    0.2 * data_set["LongPassing"]
)

print(data_set[["Name", "Score"]].head(10))
```

```python
# Ex. 16
plt.figure(figsize=(10,6))
sns.scatterplot(data=data_set, x="Wage", y="Value")
plt.title("Relatia dintre Wage si Value")
plt.show()
```
