```python
# Exercitiu 2
# Scrieți un program care cere utilizatorului să introducă nota unui examen și afișează calificativul conform regulilor:
# 9-10: Excelent
# 7-8: Bine
# 5-6: Suficient
# Sub 5: Reexaminare
# Dacă utilizatorul introduce o notă invalidă (ex. -3 sau 11), cereți-i să reintroducă o valoare validă.


while True:
    nota = int(input("Introduceti o nota: "))

    if nota < 0 or nota > 10:
        print("Nota invalida")
        continue

    match nota:
        case _ if 9 <= nota <= 10:
            print("Excelent")
        case _ if 7 <= nota <= 8:
            print("Bine")
        case _ if 5 <= nota <= 6:
            print("Suficient")
        case _:
            print("Reexaminare")

    break
```

```python
# Exercitiul 3
#Scrieți un program care generează un număr aleatoriu între 1 și 50. Utilizatorul trebuie să ghicească numărul, 
#iar programul oferă feedback:
#"Numărul este mai mare!" dacă utilizatorul a introdus un număr prea mic.
#"Numărul este mai mic!" dacă utilizatorul a introdus un număr prea mare.
#"Felicitări! Ai ghicit!" dacă numărul a fost găsit.

import random

random_number = random.randint(1, 51)
guess_number = 0

while True:
    user_number = int(input("Introdu un numar pe care sa-l ghicesti: "))
    if user_number == random_number:
        print(f"Felicitari ai ghicit numarul {random_number} din {guess_number} incercari!")
        break
    elif user_number < random_number:
        print("Numarul este mai mare!")
        guess_number += 1
    elif user_number > random_number:
        print("Numarul este mai mic!")
        guess_number += 1
```

```python
# Exercitiul 5
#Un utilizator participă la o loterie în care trebuie să aleagă un set de numere, iar programul generează numere câștigătoare.
#Utilizatorul își alege 6 numere între 1 și 49.
#Programul generează 6 numere aleatorii (simulând extragerea loteriei).
#Se compară numerele alese cu cele extrase.
#Se afișează câte numere a ghicit și un mesaj de câștig.

import random

numere_utilizator = []
for i in range(6):
    while True:
        try:
            nr = int(input(f"Introdu nr. {i+1}: "))
            if nr < 1 or nr > 49:
                print("Alege un numar de la 1 la 49!!!!!")
            elif nr in numere_utilizator:
                print("Ai ales deja acest numar!!!")
            else:
                numere_utilizator.append(nr)
                break
        except ValueError:
            print("Introdu un numar valid te rog!")

numere_castig = random.sample(range(1, 50), 6)
print("Numerele tale: ", numere_utilizator)
print("Numere castigatoare: ", numere_castig)
castigate = set(numere_utilizator) & set(numere_castig)
nr_castigate = len(castigate)

print(f"Felicitari! Ai ghicit {nr_castigate} numere: {castigate}")
if nr_castigate == 6:
    print("Felicitari pentru marele premiu!")
elif nr_castigate == 3:
    print("Felicitari ai castigat un premiu mai mic asa sa zic!")
else:
    print("Data viitoare..")
```

```python
# Exercitiul 7

#Creează un program care analizează tonul unui comentariu introdus de utilizator.
#Dacă un comentariu conține cuvinte pozitive (["bine", "frumos", "super", "excelent", "minunat"]), afișează "Comentariu pozitiv!".
#Dacă conține cuvinte negative (["urât", "prost", "groaznic", "dezamăgitor"]), afișează "Comentariu negativ!".
#Dacă nu se detectează niciun sentiment clar, afișează "Comentariu neutru.".

def user_input():
    user_text_input = input("Introduceti un text: ")
    return user_text_input

def positive_words():
    pos_words = ["bine", "frumos", "super", "excelent", "minunat"]
    return pos_words

def negative_words():
    neg_words = ["urat", "prost", "groaznic", "dezamagitor"]
    return neg_words

def final_results():
    text = user_input().lower()
    pos_word = positive_words()
    neg_word = negative_words()

    for word in pos_word:
        if word in text:
            print("Comentariu pozitiv!")
            return

    for word in neg_word:
        if word in text:
            print("Comentariu negativ!")
            return

    print("Comentariu neutru")

final_results()
```
